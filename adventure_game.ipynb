```python
# Define the rooms in the UFO as a dictionary.
# Map of the Rooms:
#   [ Room 1 ] — [ Room 2 ] — [ Room 3 ]
#       |             |             |
#   [ Room 4 ] — [ Room 5 ] — [ Room 6 ]

rooms = {
    1: {"description": "You are in the UFO Command Center. Control panels flicker with alien symbols.", 
        "north": None, "south": 4, "east": 2, "west": None},                   # Room 1 leads south to Room 4 and east to Room 2
    2: {"description": "You are in the Alien Specimen Lab. Strange creatures float in glowing tanks.", 
        "north": None, "south": 5, "east": 3, "west": 1},                      # Room 2 leads south to Room 5, east to Room 3, and west to Room 1
    3: {"description": "You are in the Navigation Room. A holographic map of galaxies spins slowly.", 
        "north": None, "south": 6, "east": None, "west": 2},                   # Room 3 leads south to Room 6 and west to Room 2
    4: {"description": "You are in the Energy Core Room. A massive, pulsating crystal hums with power.", 
        "north": 1, "south": None, "east": 5, "west": None},                   # Room 4 leads north to Room 1 and east to Room 5
    5: {"description": "You are in the Cryo-Chamber. Pods line the walls, each holding a frozen alien.", 
        "north": 2, "south": None, "east": 6, "west": 4},                      # Room 5 leads north to Room 2, east to Room 6, and west to Room 4
    6: {"description": "You are in the UFO Hangar Bay. A smaller craft is docked, ready for takeoff.", 
        "north": 3, "south": None, "east": None, "west": 5}                    # Room 6 leads north to Room 3 and west to Room 5
}



###Let's look at just Room 1 in detail:

# This tells us:  
# 1) The room has a description that will be shown to the player
# 2) The room's connections to other rooms:

# "north": None means you can't go north (None means nothing/no room)
# "south": 4 means going south takes you to Room 4
# "east": 2 means going east takes you to Room 2
# "west": None means you can't go west
``
# The game starts in Room 1 (UFO Command Center) which will keep running until the player quits.
current_room = 1

print("Welcome to the UFO Adventure! Type 'quit' to exit.")

# Start the game loop
while True:
    # Get the current room's details from the dictionary using the current_room variable.
    room = rooms[current_room]
    
    # Print the description of the current room
    print("\n" + room["description"])

    # Ask for the player's next move
    # Input is converted to lowercase so it matches the keys in the dictionary (e.g., "north", "south").
    command = input("Where do you want to go? (north/south/east/west): ").lower()

    # Check if player wants to quit
    if command == "quit":
        print("Thanks for playing!")
        break                  # Exit the loop and end the game.

    # Check if it's a valid direction in the current room.
    if command in ["north", "south", "east", "west"]:
        # Try to move in that direction
        next_room = room[command]
         # If the next_room is valid (not None), move to that room.
        if next_room is not None:
            current_room = next_room
        
        # If there is no room in the specified direction, print an error message.
        else:
            print("You can't go that direction!")
    
    # If the player's command is invalid (not a direction or quit), print an error message.   
    else:
        print("Invalid command. Use north, south, east, west, or quit.")
```
